```python
import sys

# Defining dictionary for each type of question
questions = {
			"easy":{},
			"medium":{},
			"hard":{}
			}

# Function to provide sum of the marks and their possible combinations
def subset_sum(numbers, target, sub, partial=[]):
    s = sum(partial)

    # check if the partial sum is equals to target
    if s == target:
        sub.append(partial)
    
    if s >= target:
        return  # if we reach the number why bother to continue

    for i in range(len(numbers)):
        n = int(numbers[i])
        remaining = numbers[i+1:]
        subset_sum(remaining, target, sub, partial + [n])

# Defining Subset using all the marks and Target marks
def subset(sub, numbers, target):
    sub = []
    subset_sum(numbers, target, sub, partial=[])
    return sub

# Taking input for Total Questions
count_questions = int(input("total_questions: ", ))
if count_questions <=0 or type(count_questions) != int:
    print('Atleast give 1 question in the paper')
    sys.exit()

# Providing Instructions to the user
print("Please provide question details in comma-separated format")
print("For e.g.: Q1, easy, 2")
print("where Q1 is question number")
print("easy is difficulty level")
print("and 2 is marks assigned for the question")


# Analysing all the inputs from the user
for i in range(int(count_questions)):
    ques = input().lower()
    
    try:    
        ques_number, difficulty, marks = ques.split(',')

        if "easy" in ques:
            questions['easy'][ques.split(",")[0]] = ques.split(",")[-1]
        elif "medium" in ques:
            questions['medium'][ques.split(",")[0]] = ques.split(",")[-1]
        elif "hard" in ques:
            questions['hard'][ques.split(",")[0]] = ques.split(",")[-1]
    except:
        print('Use correct format as provided : Q1, easy, 2')

# Providing Instructions to the user
print("Please provide paper details in comma-separated format")
print("For e.g.: 20, easy 25, medium 50, hard 25")
print("where 20 is total_marks")
print("easy is difficulty level followed by percentage of questions in paper")
print("medium is difficulty level followed by percentage of questions in paper")
print("difficult is difficulty level followed by percentage of questions in paper")

#Getting all the paper details
paper_details = input("paper_details: ", )

#print(questions)

#Splitting Total Marks from the input
total_marks = int(paper_details.split(",")[0])

try:
    for i in paper_details.split(","):
        if "easy" in i.lower():
            marks_easy = int(i.split(" ")[-1])*total_marks/100
        elif "medium" in i.lower():
            marks_medium = int(i.split(" ")[-1])*total_marks/100
        elif "hard" in i.lower():
            marks_hard = int(i.split(" ")[-1])*total_marks/100
except:
    print('Difficulty level not defined')
    sys.exit()

#Calculating all the values for each type of questions and giving the combination of question to the user

# EASY
try:    
    subs_easy = subset([],list(questions['easy'].values()),marks_easy)[0]

except:
    print('List index out of range as combination of "EASY" marks from the questions cannot be produced')
    sys.exit()

for key in questions['easy']:
    if int(questions['easy'][key]) in subs_easy:
        subs_easy.remove(int(questions['easy'][key]))
        print(key)

#MEDIUM
try:
    subs_medium = subset([], list(questions['medium'].values()),marks_medium)[0]
except:
    print('List index out of range as combination of "MEDIUM" marks from the questions cannot be produced')
    sys.exit()
    
for key in questions['medium']:
    if int(questions['medium'][key]) in subs_medium:
        subs_medium.remove(int(questions['medium'][key]))
        print(key)

#HARD
try:
    subs_hard = subset([], list(questions['hard'].values()),marks_hard)[0]
except:
    print('List index out of range as combination of "HARD" marks from the questions cannot be produced')
    sys.exit()

for key in questions['hard']:
    if int(questions['hard'][key]) in subs_hard:
        subs_hard.remove(int(questions['hard'][key]))
        print(key)
```
